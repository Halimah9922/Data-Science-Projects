```python
def get_percent_nan(dataframe,n=50):
  """
  This function returns a dictionary showing the features with more than n% of missing value and the exact percentage of missing values.
  """
  high_missing_value = {}

  for column in dataset.columns:
      if ((dataset[column].isnull().sum()) * (100/373008)) > n:
        high_missing_value[column] = (dataset[column].isnull().sum()) * (100/373008)
  
  return high_missing_value

```
